# 3.1 Sadece gürültü şikayetlerini seçmek

Hangi kasabanın en fazla ses şikayeti olduğunu bilmek istiyoruz. Önce nasıl görünüyor diye veriye bakalım.

```python
complaints[:5]
```

Gürültü şikayetlerini almak için "Complaint Type" sütünu "Noise - Street/Sidewalk" olan satırlar lazım. Önce nasıl yapıldığına bakalım.

```python
noise_complaints = complaints[complaints['Complaint Type'] == "Noise - Street/Sidewalk"]
noise_complaints[:3]
```

Eğer `noise_complaints`lere bakarsanız çalıştığını görürsünüz. Sadece doğru şikayet türündeki şikayetleri içeriyor. Peki bu nasıl çalışıyor? İki parçaya ayırararak inceleyelim.

```python
complaints['Complaint Type'] == "Noise - Street/Sidewalk"
```

Bu `True` ve `False` lardan oluşan büyük dizi, her satır için bir kayıt var. Data frameimizi bu array ile indexlediğimiz zaman sadece değeri `True` olan satırlar gelecektir. Filtreleme için kullanılan dizi ile data frame'in satır sayısı aynı olmalıdır.

Aynı zamanda birden fazla operatörü `&` operatörü ile birleştirebilirsiniz.

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
in_brooklyn = complaints['Borough'] == "BROOKLYN"
complaints[is_noise & in_brooklyn][:5]
```

Eğer sadece birkaç sütun ile ilgileniyorsak:

```python
complaints[is_noise & in_brooklyn][['Complaint Type', 'Borough', 'Created Date', 'Descriptor']][:10]
```

# 3.2 numpy dizileri (array) ile ilgili...

İçeride bir sütunun türü `pd.Series` dir.d.Series`

# 3.3 Öyleyse en çok ses şikayetini kim yaptı?

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
noise_complaints = complaints[is_noise]
noise_complaints['Borough'].value_counts()
```

Manhattan'mış! Ama peki bunu toplam şikayet sayısına bölersek. Bunu yapmak da kolay:

```python
noise_complaint_counts = noise_complaints['Borough'].value_counts()
complaint_counts = complaints['Borough'].value_counts()
```

```python
noise_complaint_counts / complaint_counts
```

Eğer 0 ile karşılaşırsanız bu Python 2 deki integer bölmesinden dolayıdır. düzeltmek için `complaint_counts` 'ın türünü floata çevirebiliriz.

```python
noise_complaint_counts / complaint_counts.astype(float)
```

```python
(noise_complaint_counts / complaint_counts.astype(float)).plot(kind='bar')
```

So Manhattan really does complain more about noise than the other boroughs! Neat.
